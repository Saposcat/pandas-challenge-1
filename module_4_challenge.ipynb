```python
# Create a column for shipping price.
# Assume a shipping price of $7 per pound
# for orders over 50 pounds and $10 per
# pound for items 50 pounds or under.

#df['total_weight'] = df['unit_weight'] * df['qty']
#df['shipping_unit_price'] = df['unit_weight'].apply(lambda x: 7 if x > 50 else 10)
#df['shipping_price'] = df['total_weight'] * df['shipping_unit_price']

# I think the directions here are a little bit wrong; either that, or I don't actually know what the shipping weight is.
# it could also be that the example present within the code is fallacious.
# For now I will just follow the directions as they are on the module assignment paper.

# ok, it looks like the total_weight is what we would actually use to determine the shipping price, rather than the unit weight, as I
# had initially understood.

df['total_weight'] = df['unit_weight'] * df['qty']
df['shipping_unit_price'] = df['total_weight'].apply(lambda x: 7 if x > 50 else 10)
df['shipping_price'] = df['total_weight'] * df['shipping_unit_price']

df.head()
```

```python
# Create a column for the total price
# using the subtotal and the shipping price
# along with a sales tax of 9.25%
df['tax'] = (df['line_subtotal'] + df['shipping_price']) * 0.0925
df['line_price'] = df['line_subtotal'] + df['shipping_price'] + df['tax']

df.head()
```

```python
# Create a column for the cost
# of each line using unit cost, qty, and
# shipping price (assume the shipping cost
# is exactly what is charged to the client).
df['line_cost'] = df['unit_cost'] * df['qty'] + df['shipping_price']
df.head()
```

```python
# Create a column for the profit of
# each line using line cost and line price

df['line_profit'] = df['line_price'] - df['line_cost']
df.head()
```

## Part 4: Summarize and Analyze
Use the new columns with confirmed values to find the following information.

```python
# How much did each of the top 5 clients by quantity
# spend? Check your work from Part 1 for client ids.
top_client_spends = [df.loc[df['client_id'] == client, 'line_price'].sum() for client in top_clients]

for i in range(0,5):
  print(top_clients[i],top_client_spends[i])
```

```python
# Create a summary DataFrame showing the totals for the
# for the top 5 clients with the following information:
# total units purchased, total shipping price,
# total revenue, and total profit. Sort by total profit.
top_client_df = df[df['client_id'].isin(top_clients)]
aggregated_top_df = top_client_df.groupby('client_id').agg({'qty': 'sum', 'shipping_price': 'sum',
                                                            'line_price': 'sum', 'line_profit': 'sum'})

aggregated_top_df.columns = ["units_purchased","total_shipping_price","total_revenue","total_profit"]
#funny, it looks like client ID does not count as a column as it is used to aggregate
aggregated_top_df
```

```python
# Format the data and rename the columns
# to names suitable for presentation.
# Currency should be in millions of dollars.

aggregated_top_df.columns = ["Units Purchased","Total Shipping Price","Total Revenue","Total Profit"]
#funny, it looks like client ID does not count as a column as it is used to aggregate
presentation_df = aggregated_top_df.rename_axis("Client ID").reset_index() #this seems to fix the issue

for i in [2,3,4]:
  presentation_df.iloc[:,i]  = (presentation_df.iloc[:,i] / 1000000).apply(lambda x: f'${x:.2f}MM')

presentation_df

```

```python
# Sort the updated data by "Total Profit" form highest to lowest
df_sorted = presentation_df.sort_values('Total Profit', ascending=False)
df_sorted
```
